```python
ls_features = features.name.tolist()
```

```python
sentences = []
for f in ls_features:
#     f = ''.join([i for i in f if not i.isdigit()])
    words = f.split(' ')
#     words = re.findall(r'\b[^a-zA-Z ]\w+', ' '.join(words)) # убрать английский слова
#     if not words:
#         continue
    temp = []
    for w in words:
#         if len(w) > 3:
        temp.append(w)
#     if not(not temp):
    sentences.append(temp)
```

```python
sentences
```

```python
#len(sentences)
# параметры модели Word2Vec
num_features = 16   # размерность пространства векторов (чем больше, тем более точная модель)                      
min_word_count = 10   # учитываем в модели только слова, которые встретились более, чем min_word_count раз                        
num_workers = 5      # сколько процессов одновременно будут строить модель #????????????????? # потоки
context = 10         # сколько слов в контексте следует учитывать алгоритму
downsampling = 1e-3   # исключаем очень часто встречающиеся слова

model = Word2Vec(sentences, workers=num_workers, \
            size=num_features, min_count = min_word_count, \
            window = context, sample = downsampling)

# model = Word2Vec(sentences)
```

```python
model.iter = 10
```

```python
# for epoch in range(10):
model.train(shuffle(sentences), total_examples=model.corpus_count, epochs=model.iter)
```

```python
for word, cosine in model.most_similar(positive=["магазин"], topn=10):
        print(word,cosine)
```

```python
clist = categories.category.tolist()
for c in clist:
    print('-' * 120)
    print('Категория: ', c,'\n')
    c = c.lower()
#     print(model.most_similar(positive=[c], topn=10))
    for word, cosine in model.most_similar(positive=[c], topn=25):
        print(word,cosine)
    print('-' * 120)
```
