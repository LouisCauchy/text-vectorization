```python
sentences = []
for f in list_of_features:
    f = ''.join([i for i in f if not i.isdigit()])
    words = f.split(' ')
    words = re.findall(r'\b[^a-zA-Z ]\w+', ' '.join(words)) # убрать английский слова
    if not words:
        continue
    temp = []
    for w in words:
        if len(w) > 3:
            temp.append(w)
    if not(not temp):
        sentences.append(temp)
# sentences.append(['здоровье']*50)
sentences.append(['красота'])
sentences.append(['красота',"спа","маникюр"])
sentences.append(['красота',"спа"])
sentences.append(['красота',"педикюр"])
sentences.append(['красота',"спа","салон","красоты"])
sentences.append(['здоровье',"врач","доктор"])
sentences.append(['здоровье',"врач","доктор"])
sentences.append(['кроватки',"для","младенцев"])
sentences.append(['семейные',"номера","с","детьми"])
sentences.append(['детский',"горшок"])
sentences.append(['детский',"стульчик"])
sentences.append(['детские',"игрушки"])
sentences.append(["няня","для","ребёнка"])
```

```python
#len(sentences)
# параметры модели Word2Vec
num_features = 16   # размерность пространства векторов (чем больше, тем более точная модель)                      
min_word_count = 50   # учитываем в модели только слова, которые встретились более, чем min_word_count раз                        
num_workers = 5      # сколько процессов одновременно будут строить модель #????????????????? # потоки
context = 10         # сколько слов в контексте следует учитывать алгоритму
downsampling = 1e-3   # исключаем очень часто встречающиеся слова

# model = Word2Vec(sentences, workers=num_workers, \
#             size=num_features, min_count = min_word_count, \
#             window = context, sample = downsampling)

model = Word2Vec(sentences)
```

```python
clist = categories.category.tolist()
for c in clist:
    print('-' * 120)
    print('Категоря: ', c,'\n')
    c = c.lower()
#     print(model.most_similar(positive=[c], topn=10))
    for word, cosine in model.most_similar(positive=[c], topn=10):
        print(word,cosine)
    print('-' * 120)
```

```python
members_in_cluster = round(len(model.wv.vocab) / len(categories))
members_in_cluster
```

```python
# govno
from sklearn.cluster import KMeans
import time

start = time.time() # Start time

word_vectors = model.wv.syn0
num_clusters = word_vectors.shape[0] // members_in_cluster #10

kmeans_clustering = KMeans(n_clusters = num_clusters)
idx = kmeans_clustering.fit_predict(word_vectors)

end = time.time()
elapsed = end - start
print("Time taken for K Means clustering: ", elapsed, "seconds.")
```

```python
# Create a Word / Index dictionary, mapping each vocabulary word to
# a cluster number                                                                                            
word_centroid_map = dict(list(zip(model.wv.index2word, idx)))
```

```python
for cluster in range(0,num_clusters):
    #
    # Print the cluster number  
    print("\nCluster %d" % cluster)
    #
    # Find all of the words for that cluster number, and print them out
    words = []
    for i in range(0,len(word_centroid_map.values())):
        if(list(word_centroid_map.values())[i] == cluster):
            words.append(list(word_centroid_map.keys())[i])
    print(words)
    print(len(words))
```
