```python
features['category'] = 0
```

```python
def form_categories(df, category):
    for i in range(len(df)):
        curr_feature = df['name'].iloc[i]
        for feature in category[0]:
            if feature in curr_feature:
                df.loc[i,'category'] = category[1]
    return df

for category in categ:
    form_categories(features, category)
    
for i in range(len(features)):
    curr_feature = features['category'].iloc[i]
    if curr_feature == 0:
        features['category'].iloc[i] = 'other'
```
